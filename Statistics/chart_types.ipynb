### Bar chart

```python
df_bar = pd.DataFrame({
    'categories': ['Cat 1', 'Cat 2', 'Cat 3','Cat 4', 'Cat 5'],
    'Values': [10,20,30,40,60]
})
```

```python
plt.bar(df_bar['categories'], df_bar['Values'])
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple bar chart')
plt.show()
```

```python
df_bar2 = pd.DataFrame({
    'Fruits': ['Apple','Banana','Grape','Cherries'],
    'Quantity':[4,5,8,10]
})
```

```python
fig, ax = plt.subplots()
ax.bar(df_bar2['Fruits'],df_bar2['Quantity'], color='orange')
ax.set_xlabel('Fruits')
ax.set_ylabel('Quantity')
ax.set_title('Fruits Quantity')
plt.show()
```

### Histogram

```python
# data = np.random.randint(0,100,1000)
# data = np.random.uniform(-3,3,100)
data = np.random.randn(100)
```

```python
fig, ax = plt.subplots()
ax.hist(data,bins=10, density=True)
ax.set_xlabel('Values')
ax.set_ylabel('Frequency')
ax.set_title('Histogram example')
plt.show()
```

```python
df_hist = pd.DataFrame({
    'ages': [20, 25, 20, 30, 25, 30, 35, 40, 35, 40, 45, 50, 55, 60, 55, 60, 65,20]
})
```

### Line chart

```python
df_line = pd.DataFrame({
    'Time': range(1,11),
    'Value': np.random.randint(0,10,10)
})
```

```python
fig, ax = plt.subplots()
ax.plot(df_line['Time'],df_line['Value'], color='purple')
ax.set_xlabel('Time')
ax.set_xticks(df_line['Time'])
ax.set_ylabel('Value')
ax.set_title('Line chart example')
plt.show()
```

```python
df_lines2 = pd.DataFrame({
    'Month': range(1,13),
    'Sales': [200, 250, 275, 300, 350, 400, 450, 400, 350, 300, 250, 200]
})
```

### Pie chart

```python
df_pie = pd.DataFrame({
'products': ['Product A', 'Product B', 'Product C', 'Product D'],
'sales': [15, 30, 45, 10]
})
```

```python
fig, ax = plt.subplots()
ax.pie(df_pie['sales'], labels=df_pie['products'], autopct='%1.1f%%')
ax.set_title('Product Sales')
plt.show()
```

### Area chart

```python
df_area = pd.DataFrame({
    'Month': range(1, 13),
    'Sales': [200, 250, 275, 300, 350, 400, 450, 400, 350, 300, 250, 200]
})
```

```python
fig, ax = plt.subplots()
ax.fill_between(df_area['Month'],df_area['Sales'], color='skyblue', alpha=0.3)
ax.plot(df_area['Month'],df_area['Sales'], color='darkblue')
ax.set_xlabel('Months')
ax.set_ylabel('Sales')
ax.set_title('Area chart example')
plt.show()
```

```python
df_area2 = pd.DataFrame({
    'Month': range(1, 13),
    'Product A Sales': [200, 250, 275, 300, 350, 400, 450, 400, 350, 300, 250, 200],
    'Product B Sales': [150, 200, 250, 300, 350, 400, 450, 500, 450, 400, 350, 300]
})
```

```python
fig, ax = plt.subplots()

ax.fill_between(df_area2['Month'],df_area2['Product A Sales'], color='skyblue', alpha=0.3)
ax.plot(df_area2['Month'],df_area2['Product A Sales'])
ax.fill_between(df_area2['Month'],df_area2['Product B Sales'], color='orange', alpha=0.3)
ax.plot(df_area2['Month'],df_area2['Product B Sales'])

ax.set_xlabel('Month')
ax.set_xticks(df_area2['Month'])
ax.set_ylabel('Sales')
ax.set_title('Area chart example 2')
plt.show()
```
