```python
# removendo colunas desncessarias
df.drop(['ZIP CODE','LATITUDE','LONGITUDE','LOCATION'], axis=1,inplace=True)
```

```python
# contagem por mes
month_count = df.groupby(pd.Grouper(key='CRASH DATE', freq='M')).count()
```

```python
# criando coluna com apenas os meses
df['month'] = pd.to_datetime(df['CRASH DATE']).dt.month
```

```python
df
```

```python
# refazendo a contagem por mes
month_count = df.groupby('month')['CRASH DATE'].count()
```

```python
# plotagem grafico acidentes por mes
fig, ax  = plt.subplots()
bar_container = ax.bar(month_count.index,month_count)
ax.bar_label(bar_container)
plt.show
```
