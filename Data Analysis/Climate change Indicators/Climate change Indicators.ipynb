```python
# Assuming 'df' is your DataFrame and 'Country' is the column with country names
# Also assuming that the 'Country' column contains country names exactly as they are, like 'Brazil', 'Argentina', etc.

# List of South American countries
south_america = ['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Ecuador', 'Guyana', 'Paraguay', 'Peru', 'Suriname', 'Uruguay', 'Venezuela']

# Create a new column 'Region'
df['Region'] = df['Country'].apply(lambda x: 'SA' if x in south_america else 'Other')
```

```python
df_sa = df.loc[df['Region'] == 'SA']
```

```python
df_sa[['F2000', 'F2001', 'F2002', 'F2003', 'F2004',
       'F2005', 'F2006', 'F2007', 'F2008', 'F2009', 'F2010', 'F2011', 'F2012',
       'F2013', 'F2014', 'F2015', 'F2016', 'F2017', 'F2018', 'F2019', 'F2020',
       'F2021', 'F2022']].describe()
```

```python
sa_mean_year = df_sa[['F2000', 'F2001', 'F2002', 'F2003', 'F2004',
       'F2005', 'F2006', 'F2007', 'F2008', 'F2009', 'F2010', 'F2011', 'F2012',
       'F2013', 'F2014', 'F2015', 'F2016', 'F2017', 'F2018', 'F2019', 'F2020',
       'F2021', 'F2022']].mean()
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(sa_mean_year.index, sa_mean_year.values)
ax.set_title('South America mean since 2000')
plt.xticks(rotation=90)
plt.show()
```

```python
fig, sx = plt.subplots(figsize=(12,6))

selected_years = ['F2020', 'F2021', 'F2022']

for year in selected_years:
    bars = sx.plot(df_sa['Country'], df_sa[year], label=year)
    sx.set_title('2020 to 2022 Climate Change indicator')

sx.legend()

plt.show()
```
