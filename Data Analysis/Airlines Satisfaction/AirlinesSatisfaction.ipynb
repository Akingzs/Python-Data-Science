```python
# Fazendo a media de todos os serviços para ter uma avaliação geral
df['average'] = df.iloc[:, 9:23].mean(axis=1)
```

```python
# A distância do voo afeta as preferências ou os padrões de voo do cliente?

df['average'].describe()
```

```python
bins = [0,2,3,4,5]
bins_label = ['Muito Ruim','Ruim','OK','Muito Bom']
df['Average bins'] = pd.cut(df['average'], bins=bins, labels=bins_label)

bins_count = df['Average bins'].value_counts()

fig, ax = plt.subplots()
bar_container = ax.bar(bins_count.index, bins_count)
ax.set(ylabel='Returning count', title='Returning clients per Gender')
ax.bar_label(bar_container)
```

```python
dbins_count = df['Distance bins'].value_counts()

fig, ax = plt.subplots()
bar_container = ax.bar(dbins_count.index.astype(str), dbins_count)
ax.set(ylabel='Returning count', title='Returining clients per Gender')
ax.bar_label(bar_container)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.show
```
